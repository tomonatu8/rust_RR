# 数値実験結果の表示

## (1) ラウンド数は固定で，エージェントの数を増加させる
### $\log_2 n$ との関係を調べる

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
filename_csv = '../rust_rr/incident_list.csv'
df = pd.read_csv(filename_csv,header=None)
print(df)

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(df.loc[:,0], df.loc[:,1], label='Results')
ax.plot(df.loc[:,0], df.loc[:,2], label='log_2 N')
ax.plot(df.loc[:,0],2*df.loc[:,2], label='2*log_2 N')
ax.legend()
ax.set_xlabel("N")
ax.grid(which = "major", axis = "x",  alpha = 0.8,
        linestyle = "--", linewidth = 1)

ax.grid(which = "major", axis = "y", alpha = 0.8,
        linestyle = "--", linewidth = 1)
```

## (2) エージェントド数は固定で，ラウンドの数 (財の数) を増加させる

### $n = 8$

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

n = "8"
filename_csv = '../rust_rr/' + n + '_incident_list_.csv'
df = pd.read_csv(filename_csv,header=None)
print(df)
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(df.loc[:,0], df.loc[:,1], label='Results')
ax.legend()
ax.set_title("n = " + n)
ax.set_xlabel("The number of rounds")
ax.grid(which = "major", axis = "x",  alpha = 0.8,
        linestyle = "--", linewidth = 1)

ax.grid(which = "major", axis = "y", alpha = 0.8,
        linestyle = "--", linewidth = 1)
```
